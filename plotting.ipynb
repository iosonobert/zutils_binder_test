```python
from zutils import plotting as zplot
import matplotlib.pyplot as plt
```

```python
widths=[2,4,1]     # Width of each column of axes
heights=[2,4,1]    # Height of each row of axes
hspace=[2,3]       # Horizontal spacing between columns. Use single value to set all vertical spaces evenly. 
                   # Use list to set each spacing seperately
vspace=1           # Vertical spacing between rows. Use single value to set all vertical spaces evenly

bottom = 3
top = 3

zl = zplot.axis_layer(widths=widths, heights=heights, hspace=hspace, vspace=vspace, bottom=bottom, top=top)
zl.verbose = False # Reduce printouts
```

```python
f = plt.figure(dpi=100)

zl.lay(0,0, colbleed=2)

zl.lay(1,0)
zl.lay(1,2)

zl.lay(2,0)
zl.lay(2,1, rowbleed=1)
zl.lay(2,2)

f.savefig('axis_layer_test.png')
```
